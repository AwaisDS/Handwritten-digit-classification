```python
from sklearn.datasets import load_digits
import pandas as pd
```

```python
digits=load_digits()
```

```python
X=digits.data
y=digits.target
```

```python
feature_names = [f'pixel_{i}' for i in range(X.shape[1])]

df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

print(df.head())
```

```python
df.to_csv('digits.csv')
```
